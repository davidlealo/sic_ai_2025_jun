```python
# Importaciones
import pandas as pd
import numpy as np
```

```python
# Base de datos: personas
df_personas = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Nombre': ['Alicia', 'Pedro', 'Laura', 'Tomás', 'Beatriz'],
    'Edad': [28, 35, 24, 30, 28],
    'Zona': ['Norte', 'Norte', 'Sur', 'Sur', 'Sur']  # Nueva columna: zona geográfica
})

print(" DataFrame original:")
print(df_personas)
```

```python
df_ordenado_edad = df_personas.sort_values(by='Edad')
print("\n Ordenado por Edad (ascendente):")
print(df_ordenado_edad)
```

```python
df_ordenado_edad = df_personas.sort_values(by='Edad', ascending=False)
print("\n Ordenado por Edad (ascendente):")
print(df_ordenado_edad)
```

####**¿Qué es un MultiIndex?**
Es un índice con varios niveles, por ejemplo:

- Nivel 1: Zona (como "Norte" o "Sur")

- Nivel 2: ID de la persona

>Sirve para organizar mejor los datos agrupados, como si fueran carpetas y subcarpetas.

```python
# Crea un DataFrame con personas, su nombre, edad, y zona.

# Toma la zona y el ID para formar un índice compuesto (jerárquico).

# Usa pd.MultiIndex.from_tuples() para hacer ese índice múltiple.

# Crea un nuevo DataFrame df_multi que usa ese índice jerárquico.


# Crear los niveles del índice jerárquico
nivel_1 = df_personas['Zona']       # Ej: 'Norte', 'Sur'
nivel_2 = df_personas['ID']         # ID de cada persona

# Crear tuplas y MultiIndex
multi_index = pd.MultiIndex.from_tuples(list(zip(nivel_1, nivel_2)), names=['Zona', 'ID'])

# Crear nuevo DataFrame con índice jerárquico
df_multi = pd.DataFrame(df_personas[['Nombre', 'Edad']].values, index=multi_index, columns=['Nombre', 'Edad'])

# Mostrar resultado
print(df_multi)
```
