```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

data = fetch_california_housing(as_frame=True)
df = data.frame

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df[['MedInc','AveRooms']]
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Escalador para modelos que lo requieren
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

# Modelos
lr = LinearRegression().fit(X_train, y_train)
tree = DecisionTreeRegressor(max_depth=5).fit(X_train, y_train)
knn = KNeighborsRegressor(n_neighbors=5).fit(X_train_scaled, y_train)
svm = SVR().fit(X_train_scaled, y_train)
```

```python
sample = pd.DataFrame([{
    'MedInc': 3.5,
    'AveRooms': 4
}])

pred_lr = lr.predict(sample)
pred_tree = tree.predict(sample)
pred_knn = knn.predict(scaler.transform(sample))
pred_svm = svm.predict(scaler.transform(sample))

#sample_scaled = scaler.transform(sample)  # para KNN y SVR
```

```python
print(" Regresi贸n Lineal:", round(pred_lr[0] * 100000, 2), "USD")
print(" rbol de Decisi贸n:", round(pred_tree[0] * 100000, 2), "USD")
print(" KNN:", round(pred_knn[0] * 100000, 2), "USD")
print(" SVR:", round(pred_svm[0] * 100000, 2), "USD")
```

```python
# 1. Cargar y explorar el dataset
# Importar las bibliotecas necesarias: fetch_california_housing para cargar el dataset y pandas para manipular datos
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Cargar el dataset California Housing con as_frame=True para obtener un DataFrame
data = fetch_california_housing(as_frame=True)
# Extraer el DataFrame, que contiene las columnas predictoras y la columna objetivo (MedHouseVal)
df = data.frame

# 2. Preparar los datos
# Importar herramientas para dividir los datos y escalar las caracter铆sticas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separar las variables predictoras (X) y la variable objetivo (y)
# Excluir la columna 'MedHouseVal' (valor de la vivienda) de X, usando axis=1 para indicar que es una columna
X = df.drop('MedHouseVal', axis=1)
# Asignar la columna 'MedHouseVal' como la variable objetivo y (valores de viviendas en cientos de miles de d贸lares)
y = df['MedHouseVal']

# Dividir los datos en conjuntos de entrenamiento (75%) y prueba (25%)
# random_state=42 asegura que la divisi贸n sea reproducible
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Crear un escalador para estandarizar las caracter铆sticas (restar la media y dividir por la desviaci贸n est谩ndar)
# Esto es necesario para modelos como KNN y SVR, que son sensibles a la escala de las variables
scaler = StandardScaler()
# Ajustar el escalador a los datos de entrenamiento y transformarlos (calcula media y desviaci贸n est谩ndar)
X_train_scaled = scaler.fit_transform(X_train)
# Transformar los datos de prueba usando el mismo escalador (evita filtraci贸n de datos)
X_test_scaled = scaler.transform(X_test)

# 3. Entrenar los modelos
# Importar las clases de los modelos de regresi贸n de sklearn
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

# Entrenar los cuatro modelos de regresi贸n:
# - LinearRegression: Modelo lineal que asume una relaci贸n lineal entre caracter铆sticas y el valor de la vivienda
lr = LinearRegression().fit(X_train, y_train)
# - DecisionTreeRegressor: rbol de decisiones con profundidad m谩xima de 5 para evitar sobreajuste
tree = DecisionTreeRegressor(max_depth=5).fit(X_train, y_train)
# - KNeighborsRegressor: Predice basado en los 5 vecinos m谩s cercanos; usa datos escalados
knn = KNeighborsRegressor(n_neighbors=5).fit(X_train_scaled, y_train)
# - SVR: M谩quina de soporte vectorial para regresi贸n; usa datos escalados
svm = SVR().fit(X_train_scaled, y_train)

# 4. Crear la observaci贸n de prueba
# Crear un DataFrame con una sola fila que representa una vivienda con las caracter铆sticas especificadas
# Solo MedInc (3.5, equivalente a $35,000) y AveRooms (4) son requeridos por la pregunta; los dem谩s valores son del ejemplo
sample = pd.DataFrame([{
    'MedInc': 3.5,      # Ingreso medio del vecindario ($35,000 en decenas de miles)
    'HouseAge': 25,     # Edad promedio de las casas (25 a帽os)
    'AveRooms': 4,      # Promedio de 4 habitaciones por vivienda
    'AveBedrms': 1,     # Promedio de 1 dormitorio por vivienda
    'Population': 1000, # Poblaci贸n del 谩rea (1000 personas)
    'AveOccup': 3,      # Promedio de 3 ocupantes por hogar
    'Latitude': 34,     # Latitud geogr谩fica (cerca de Los ngeles)
    'Longitude': -118   # Longitud geogr谩fica
}])

# Escalar la observaci贸n de prueba para los modelos KNN y SVR, usando el mismo escalador ajustado en X_train
sample_scaled = scaler.transform(sample)

# 5. Hacer las predicciones
# Predecir el valor de la vivienda con cada modelo y convertir el resultado a d贸lares (multiplicar por 100,000)
# Redondear a 2 decimales para mayor claridad
# Regresi贸n Lineal usa datos no escalados
print(" Regresi贸n Lineal:", round(lr.predict(sample)[0] * 100000, 2), "USD")
# rbol de Decisi贸n usa datos no escalados
print(" rbol de Decisi贸n:", round(tree.predict(sample)[0] * 100000, 2), "USD")
# KNN usa datos escalados, predice basado en los 5 vecinos m谩s cercanos
print(" KNN:", round(knn.predict(sample_scaled)[0] * 100000, 2), "USD")
# SVR usa datos escalados, predice basado en un hiperplano de regresi贸n
print(" SVR:", round(svm.predict(sample_scaled)[0] * 100000, 2), "USD")
```
